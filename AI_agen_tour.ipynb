```python
import google.generativeai as genai
from google.colab import userdata
from langchain.schema import SystemMessage, HumanMessage, AIMessage
import re
from datetime import datetime

genai.configure(api_key=userdata.get('GOOGLE_API_KEY'))
```

```python
paket_wisata = {
    "england": 25000000,
    "japan": 24000000,
    "singapore": 20000000
}

maskapai = {
    "etihad": 5000000,
    "emirates": 6000000,
    "garuda": 4500000
}

payment_methods = ["bca", "mandiri", "bri", "btn", "danamon"]
add_on_price = 500000

order_data = {
    "nama": None,
    "kontak": None,
    "tujuan": None,
    "jumlah": None,
    "tanggal": None,
    "maskapai": None,
    "payment": None,
    "add_on": None,
    "total": None,
    "konfirmasi": None
}

def valid_tanggal(tanggal):
    try:
        datetime.strptime(tanggal, "%d/%m/%Y")
        return True
    except:
        return False


def reset_order():
    for key in order_data:
        order_data[key] = None


def chatbot_response(user_input):

    forbidden_words = ["presiden", "agama", "politik", "sejarah"]
    if any(w in user_input.lower() for w in forbidden_words):
        return "Maaf, saya hanya melayani pemesanan paket wisata."

    # ========================== ALUR PEMESANAN ==============================

    if order_data["tujuan"] is None:
        for t in paket_wisata.keys():
            if t in user_input.lower():
                order_data["tujuan"] = t
                return "Boleh tahu nama Anda?"
        return "Silakan pilih tujuan: England, Japan, atau Singapore."

    if order_data["nama"] is None:
        order_data["nama"] = user_input.title()
        return f"Baik {order_data['nama']}, berapa jumlah orang?"

    if order_data["jumlah"] is None:
        if not user_input.isdigit():
            return "Jumlah harus berupa angka ya!"
        order_data["jumlah"] = int(user_input)
        return "Tanggal keberangkatan berapa? Format: DD/MM/YYYY"

    if order_data["tanggal"] is None:
        if not valid_tanggal(user_input):
            return "Format tanggal tidak valid! Gunakan DD/MM/YYYY"
        order_data["tanggal"] = user_input
        return "Maskapai yang diinginkan? Etihad / Emirates / Garuda"

    if order_data["maskapai"] is None:
        for m in maskapai.keys():
            if m in user_input.lower():
                order_data["maskapai"] = m

                tujuan = order_data["tujuan"]
                jumlah = order_data["jumlah"]
                order_data["total"] = paket_wisata[tujuan] * jumlah + maskapai[m] * jumlah

                return (
                    f"Total saat ini: Rp {order_data['total']:,}\n"
                    "Metode pembayaran? (BCA / Mandiri / BNI BTN, Danamon"
                )
        return "Maskapai tidak tersedia! Pilih Etihad / Emirates / Garuda"

    if order_data["payment"] is None:
        for pay in payment_methods:
            if pay in user_input.lower():
                order_data["payment"] = pay
                return "Apakah ingin tambah asuransi Rp 500.000 per orang? (Y/N)"
        return "Metode pembayaran tidak valid!"

    if order_data["add_on"] is None:
      if user_input.lower() == "y":
        order_data["add_on"] = True
        order_data["total"] += add_on_price * order_data["jumlah"]
        return "Nomor WhatsApp Anda?"

    elif user_input.lower() == "n":
        order_data["add_on"] = False
        return "Nomor WhatsApp Anda?"

    return "Silakan jawab Y atau N."


    if order_data["kontak"] is None:
        order_data["kontak"] = user_input
        order_data["konfirmasi"] = False

        return (
            f"‚úÖ Pemesanan Berhasil!\n\n"
            f"Nama: {order_data['nama']}\n"
            f"Tujuan: {order_data['tujuan'].title()}\n"
            f"Jumlah: {order_data['jumlah']} orang\n"
            f"Keberangkatan: {order_data['tanggal']}\n"
            f"Maskapai: {order_data['maskapai'].title()}\n"
            f"Pembayaran: {order_data['payment'].upper()}\n"
            f"Asuransi: {'Ya' if order_data['add_on'] else 'Tidak'}\n"
            f"Total Biaya: Rp {order_data['total']:,}\n"
            f"Kontak: {order_data['kontak']}\n\n"
            "Apakah Anda sudah selesai melakukan pemesanan? (Y/N)"
        )

    # ====================== KONFIRMASI AKHIR ===========================

    if order_data["konfirmasi"] is False:
      if user_input.lower() == "y":
        print("Terima kasih telah memesan di Celerates Tour Agency! ‚úàÔ∏èüòä")
        exit()  # ‚úÖ Program berhenti total

    elif user_input.lower() == "n":
        reset_order()
        return (
            "Baik! Silakan pilih kembali tujuan wisata:\n"
            "- England\n- Japan\n- Singapore"
        )

    return "Silakan jawab dengan Y atau N."
```

```python
# Running / Test
print("Bot : Selamat datang di Celerates Tour Agency! "
      "Tujuan wisata mana yang Anda inginkan? (England / Japan / Singapore)")
chat_history = []
while True:
    user_input = input("Anda : ")
    response = chatbot_response(user_input)
    chat_history.append(response)
    print("Bot :", response)
```
